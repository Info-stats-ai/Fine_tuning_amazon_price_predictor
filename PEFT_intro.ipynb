### When you see 4096 in the lora_q_proj calculation (e.g., 4096 * 32), it refers to the in_features or out_features of the linear layer (q_proj in this case) that the LoRA adapter is modifying. Since the q_proj layer itself operates on input/output dimensions of 4096, these are the dimensions that are being reduced to 32 (your LORA_R or rank) and then expanded back.

So, for a layer with input_dim = 4096 and output_dim = 4096 (like q_proj):

LoRA_A matrix has dimensions input_dim x rank (i.e., 4096 x 32)
LoRA_B matrix has dimensions rank x output_dim (i.e., 32 x 4096)
The calculation 4096 * 32 + 4096 * 32 correctly sums the parameters for both lora_A and lora_B for that specific projection. The 4096 comes from the internal feature dimension of the model, which is consistently used across many layers, including the embedding output.

```python
# Each of the Target Modules has 2 LoRA Adaptor matrices, called lora_A and lora_B
# These are designed so that weights can be adapted by adding alpha * lora_A * lora_B
# Let's count the number of weights using their dimensions:

# See the matrix dimensions above
lora_q_proj = 4096 * 32 + 4096 * 32
lora_k_proj = 4096 * 32 + 1024 * 32
lora_v_proj = 4096 * 32 + 1024 * 32
lora_o_proj = 4096 * 32 + 4096 * 32

# Each layer comes to
lora_layer = lora_q_proj + lora_k_proj + lora_v_proj + lora_o_proj

# There are 32 layers
params = lora_layer * 32

# So the total size in MB is
size = (params * 4) / 1_000_000

print(f"Total number of params: {params:,} and size {size:,.1f}MB")
```

### What is this adapter?

In the context of Parameter-Efficient Fine-Tuning (PEFT) techniques like LoRA (Low-Rank Adaptation), an adapter refers to a small, trainable neural network component that is added to a pre-trained large model (the base_model). Instead of fine-tuning all the parameters of the massive base model, you only train these much smaller adapter modules.

Specifically, a LoRA adapter consists of two small matrices, often called lora_A and lora_B, which are injected into the key layers of the pre-trained model (like the query, key, value, and output projection matrices in the attention mechanism). During training, only lora_A and lora_B are updated, while the original weights of the base model remain frozen.

When the model is used (for inference or further training), the outputs of lora_A and lora_B are multiplied and then added to the output of the original pre-trained layer. This effectively modifies the behavior of the original layer in a parameter-efficient way. The 109.1MB represents the total size of all these lora_A and lora_B matrices across the specified layers of your model.
